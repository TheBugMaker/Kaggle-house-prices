```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
pd.set_option('display.max_columns', 500)
%matplotlib inline
```

```python
df = pd.read_csv('house_data/train.csv')
df.head()
```

```python
df =  df.drop(['Id'],axis=1)
```

```python
df['SalePrice'].groupby(df['YrSold']).describe() 
```

```python
var = 'OverallQual'
data = pd.concat([df['SalePrice'], df[var]], axis=1)
f, ax = plt.subplots(figsize=(8, 6))
fig = sns.boxplot(x=var, y="SalePrice", data=data)
fig.axis(ymin=0, ymax=800000);
```

```python
numeric_cols = [cname for cname in df.columns if 
                                df[cname].dtype in ['int64', 'float64']]
df[numeric_cols].describe()
```

```python
catagorical_cols = [cname for cname in df.columns if
                                df[cname].dtype == "object"]
df[catagorical_cols].describe()
```

```python
train = pd.get_dummies(df[catagorical_cols+numeric_cols])
train.head()
```
